# Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

# Load Data

```python
df = pd.read_csv("../data/gdp_cleaned.csv")
```

# Top 10 Economies

```python
latest_year = df["Year"].max()
latest_df = df[df["Year"] == latest_year]
top10 = latest_df.nlargest(10, "GDP_Billions")
```

# Create a chart: Bar chart

```python
plt.figure(figsize = (12,6))
plt.bar(top10["country"], top10["GDP_Billions"])
plt.xticks(rotation = 45)
plt.title("Top 10 Economics by GDP")
plt.show()
```
