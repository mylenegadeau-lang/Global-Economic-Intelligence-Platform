# Question 1: How many unique entities (countries/aggregates) are in the dataset?

```python
total_entities = df["country"].nunique()
print(f"Total unique country/regional entries: {total_entities}")
```

# Question 2: What is the range of years covered in the dataset?

```python
min_year = int(df["Year"].min())
max_year = int(df["Year"].max())
print(f"Earliest year: {min_year}")
print(f"Latest year: {max_year}")
print(f"Total span: {max_year - min_year + 1} years")
```

# Question 3: How much missing data (null values) exists?

```python
missing_df = pd.DataFrame({"Missing Values": df.isnull().sum(), "Percentage (%)": (df.isnull().mean() * 100).round(2),})
missing_df
```

# Question 4: Top Economies in the Latest Available Year

```python
latest_year = int(df["Year"].max())
df_latest = df[(df["Year"] == latest_year) & (df["GDP_Billions"].notnull())]

#Show top 10 entities by GDP
top_10 = df_latest.sort_values(by="GDP_Billions", ascending=False).head(10)
top_10[["country", "country_code", "Year", "GDP_Billions"]]
```
